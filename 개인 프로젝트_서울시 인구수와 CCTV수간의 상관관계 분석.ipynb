```python
import pandas as pd
cctv = pd.read_csv('CCTV_in_Seoul.csv', encoding = 'utf-8')
cctv
```

```python
cctv.rename(columns = {cctv.columns[0] : '자치구', 
                       cctv.columns[1] : 'CCTV수'}, inplace = True)
cctv.head()
```

```python
cctv['CCTV수 증가율'] = (cctv['2016년'] + cctv['2015년'] + cctv['2014년']) / cctv['2013년도 이전'] * 100
cctv.head()
```

```python
del cctv['2013년도 이전']
del cctv['2014년']
del cctv['2015년']
del cctv['2016년']
cctv.head()
```

```python
population = pd.read_excel('population_in_Seoul.xls',
                          header = 2, # 엑셀의 세번째 행부터 읽어옴
                          usecols = 'B, D, G, J, N') # B,D,G,J,N열을 읽어옴
population
```

```python
population.rename(columns={population.columns[1] : '인구수',
                          population.columns[2] : '한국인',
                          population.columns[3] : '외국인',
                          population.columns[4] : '고령자'}, inplace = True)
population.head()
```

```python
df = pd.merge(cctv, population, on='자치구')
df.head()
```

```python
df.set_index('자치구', inplace=True)
df.head()
```

```python
H0 = "인구수와 CCTV수는 상관관계가 없다"
H1 = "인구수와 CCTV수는 상관관계가 있다"

print("귀무가설은 ", H0)
print("대립가설은 ", H1)
```

```python
import numpy as np

# (전국CCTV수)모평균 구하기
#2016년: 105904 / 2015년: 84202, 2014년: 89307
#2013년이전: 256496
all_mean = (105904 + 84202 + 89307) / 256496 * 100
print(all_mean)

# 표본평균
cctv_mean = np.mean(df['CCTV수'])
print(cctv_mean)
# 표준편차
cctv_std = np.std(df['CCTV수'])
print(cctv_std)
# n
sample_n = 25

sample_error = cctv_mean - all_mean
SE = cctv_std / np.sqrt(sample_n)

# 검정 통계량 Z
z = sample_error / SE

print("검정통계량 Z는 ", z)
```

```python
if abs(z) <= 1.96 :
    print("귀무가설 \"", H0, "\"", "-> 기각 실패")
else :
    print("귀무가설 \"", H0, "\"", "-> 기각 성공")
```

```python
from scipy import stats

# 정규분포 그리기 
z_dist = stats.norm(loc = 0, scale = 1) # 평균이0이고 표준편차가 1인 정규분포
x = np.linspace(-2, 2, 100)
y = z_dist.pdf(x)

plt.plot(x, y)
plt.title("PDF of z-distribution")

# 오른쪽
right = np.linspace(abs(z), x[-1], 50)
plt.fill_between(right, z_dist.pdf(right), color = "pink")

# 왼쪽
left = np.linspace(-x[-1], -abs(z), 50)
plt.fill_between(left, z_dist.pdf(left), color = "pink")

p = z_dist.sf(abs(z))
p = p * 2
print("p value는", p)

if p >= 0.05 :
    print("결과:", H0, "-> 귀무가설 기각 실패")
    
else :
    print("결과:", H1, "-> 귀무가설 기각 성공")
```

```python
# 예측모델 만들기: x축-'인구수', y축-'CCTV수', 모델의 차수-'1차'
fp1 = np.polyfit(df['인구수'], df['CCTV수'], 1)

# poly1d()함수를 이용하여 fp1 모델을 다항식으로 변환하고, 그래프 그리기
f1 = np.poly1d(fp1)

# 10만부터 70만까지 100의 구간으로 나눠 그래프 그리기
fx = np.linspace(100000, 700000, 100)

# 오차값(종속변수의 변동정도)을 계산하여, df에 '오차' column을 추가
df['오차'] = np.abs(df['CCTV수'] - f1(df['인구수']))

# 오차값을 기준으로 내림차순 정렬
df_sort = df.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(df['인구수'], df['CCTV수'],
           c=df['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='purple')

for n in range(25):
    plt.text(df_sort['인구수'][n]*1.02, df_sort['CCTV수'][n]*0.98,
             df_sort.index[n], fontsize=15
            )

plt.xlabel('인구수')
plt.ylabel('인구수에 따른 CCTV수')
plt.colorbar()
plt.grid()
plt.show()
```
